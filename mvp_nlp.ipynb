```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_excel("prueba.xlsx")
```

```python
df
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(df["texto"])
X
```

```python
X_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
X_df
```

```python
word_counts = X.sum(axis=0)
word_counts
```

```python
word_frecuency = [(word, int(word_counts[0, idx])) for word, idx in vectorizer.vocabulary_.items()]
word_frecuency
```

```python
word_frecuency = sorted(word_frecuency, key=lambda x: x[1], reverse=True)
```
